## Generation of Complex Exponential 
freq = 400, sample_rate = 8000 

```
import numpy as np
import matplotlib.pyplot as plt
```

```
fig_size = (7, 5)
fig_size_half = (fig_size[0], fig_size[1] / 2.0)
```

```
sample_rate = 8000
freq = 400
phi = 30
amp = 1.2
time_total = 1.0
```

```
time = np.arange(np.round(time_total * sample_rate)) / float(sample_rate)
complex_exp = amp * np.exp(1j * (2 * np.pi * freq * time + (np.pi/(180 * phi))))
```

```
plt.figure(1, figsize=fig_size)
plt.plot(time[0:48], complex_exp.real[0:48], '-b', label='complex.real')
plt.plot(time[0:48], complex_exp.imag[0:48], '--r', label='complex.real')
plt.ylim([-1.5, 1.5])
plt.ylabel('Real, Imaginary')
plt.xlabel('time [sec]')
title = 'Real/Imaginary Parts of complex Exponential $x(t)$'
title = title + f'$f_m={freq}$ Hz, $F_s={sample_rate}$'
plt.title(title)
plt.legend(loc=1)
plt.grid() # comente and view
plt.show()
```

```
# make stem plot
plt.figure(1, figsize=fig_size)
plt.stem(time[:24], complex_exp[:24])
plt.ylabel("$x(t)$")
plt.xlabel("t [sec]")
title = "Stem Plot of Sinusoidal Waveform $x(t)$"
title = title + f", $f_m={freq}$ Hz, $F_s={sample_rate}$ Hz"
plt.title(title)
plt.grid()
plt.show()
```

## Sobre a relação entre taxa de amostragem e frequência

Podemos perceber que a melhor relação para a plotagem da senóide a partir da exponencial complexa é quando utilizamos a frequência de amostragem pelo menos em 20 vezes a frequência da senoíde.

##Gerando um sinal PAM

```
percent = 40.0
time_period = 10.0
cycles = 30
dt = 0.01
```

```
t = np.arange(0, cycles * time_period, dt)
pwm = (t % time_period) < (time_period * percent/100)
x = np.linspace(-10, 10, len(pwm))

y = np.sin(x)

plt.plot(t,y)
plt.ylim([y.min(), y.max()])
plt.show()
```

```
y[pwm == 0] = 0

plt.plot(t,y)
plt.ylim([y.min(), y.max()])
plt.grid()
plt.show() 
```
